```python
#distribution plots

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
#People who left the tip after the meal and other metadata
```

```python
#dist plot 

sns.distplot(tips['total_bill'],kde=False,bins=30)

#the line below is the 'kde'=Kernel Density Estimation
```

Most of the tips happen between 10 to 30 dollars. we can change the binsize to view granular level of details.
Kde can be True if we want to see the line


```python
#Can be used for bivariate data (more than two data variables)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')   #hex, reg
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#Build kde plot
sns.kdeplot(tips['total_bill'])


```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True)
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths=0.3)
```

```python
sns.clustermap(fp,cmap='magma')    # ,standard_scale=1
```

Grids

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

```

```python
h=sns.FacetGrid(data=tips,col='time',row='smoker')

h.map(sns.distplot,'total_bill')
```

```python
h=sns.FacetGrid(data=tips,col='time',row='smoker')

h.map(plt.scatter,'total_bill','tip')
```

Regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',  markers=['o','v'],
          scatter_kws={'s':100}
          )
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.7,size=8)
```
