Matplot lib


```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x=np.linspace(0,5,11)
y=x**2
```

```python
#functinal method

plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')


plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_xlabel('Y label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])  #left bottom width height


axes1.plot(x,y)
axes2.plot(x,y)
```

```python
#subplots

fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('Text')

axes[1].plot(y,x)
```

Figsize and DPI(Dots per inch)


```python
fig =plt.figure(figsize=(8,2))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
fig.savefig
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x2')
ax.plot(x,x**3,label='x3')

ax.legend(loc='best')
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth=3,linestyle=':',marker='o',markerfacecolor='blue',markersize=10)
```
